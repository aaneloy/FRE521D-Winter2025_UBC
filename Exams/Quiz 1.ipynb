## Run this cell and load the data (Do not change anything in this cell)

```python
import numpy as np
from sklearn.datasets import load_wine
import pandas as pd

# Load the Wine dataset from sklearn
wine_data = load_wine(as_frame=True)  # Load the dataset as a Pandas DataFrame
wine_df = wine_data.frame  # Access the DataFrame

wine_df.rename(columns={'target': 'class'}, inplace=True)
wine_df['quality_category'] = pd.cut(
    wine_df['class'],  
    bins=[-1, 0.5, 1.5, 2.5],  
    labels=['Low Quality', 'Medium Quality', 'High Quality']
)
np.random.seed(42)
m = 0.3
total_cells = wine_df.size
num = int(total_cells * m)
random_rows = np.random.randint(0, wine_df.shape[0], size=num)
random_cols = np.random.randint(0, wine_df.shape[1], size=num)
for row, col in zip(random_rows, random_cols):
    wine_df.iloc[row, col] = np.nan

wine_df.head()
```
